```python
import spacy
import re
from spacy.matcher import Matcher
nlp = spacy.load("en_core_web_sm")
matcher = Matcher(nlp.vocab)
```

```python
import os
fileids = os.listdir("texts")
fileids = [("texts/" + fileid) for fileid in fileids]
fileids
```

```python
textlist = []
for fileid in fileids:
    with open (fileid, mode ="r", encoding = "utf-8") as f:
        text = f.read()
        textlist.append(text)
textstring = str(textlist)
```

```python
pipeline = [('“', '"'), ('´´', '"'), ('”', '"'), ('’’', '"')]
for old, new in pipeline:
    textstring = textstring.replace(old, new)
```

```python
textstring = textstring.replace("\\n", "")
textstring = textstring.replace("\'", "")
textstring
```

```python
matcher = Matcher(nlp.vocab)
pattern_n = [{"POS": "PROPN"}]
matcher.add("PROPER_NOUNS", [pattern_n], greedy="LONGEST")
doc = nlp(textstring)
matches = matcher(doc)
print (len(matches))
for match in matches[:1000]:
    print (match, doc[match[1]:match[2]])
    
```

```python
matcher = Matcher(nlp.vocab)
pattern_q = [{'ORTH': '"'}, {'IS_ALPHA': True, "OP": "+"}, {'IS_PUNCT': True, "OP": "*"}, {'ORTH': '"'}]
matcher.add("QUOTES", [pattern_q], greedy='LONGEST')
doc = nlp(textstring)
matches_q = matcher(doc)
matches_q.sort(key = lambda x: x[1])
print (len(matches_q))
for match in matches_q[:100]:
    print (match, doc[match[1]:match[2]])
    with open ("approach2output.txt","a") as x:
        print(match, doc[match[1]:match[2]], file = x)
```
