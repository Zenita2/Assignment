```python
import spacy
import re
from spacy.matcher import Matcher
nlp = spacy.load("en_core_web_sm")
matcher = Matcher(nlp.vocab)
```

```python
import os
fileids = os.listdir("texts")
fileids = [("texts/" + fileid) for fileid in fileids]
fileids
```

```python
textlist = []
for fileid in fileids:
    with open (fileid, mode ="r", encoding = "utf-8") as f:
        text = f.read()
        textlist.append(text)
textstring = str(textlist)
```

```python
pipeline = [('“', '"'), ('´´', '"'), ('”', '"'), ('’’', '"')]
for old, new in pipeline:
    textstring = textstring.replace(old, new)
```

```python
textstring = textstring.replace("\\n", "")
textstring = textstring.replace("\'", "")
```

```python
textstring
```

```python
def find_sents(text):
    doc = nlp(text)
    sentences = list(doc.sents)
    return(sentences)
```

```python
def get_quotes(text):
    quotes = re.findall(r' "(.*?)"', text)
    return(quotes)
```

```python
found_sents = find_sents(textstring)
```

```python
found_sents
```

```python
matcher = Matcher(nlp.vocab)
pattern_n = [{"POS": "PROPN"}]
matcher.add("PROPER_NOUNS", [pattern_n], greedy="LONGEST")
doc = nlp(text)
matches = matcher(doc)
print (len(matches))
for match in matches[:10]:
    print (match, doc[match[1]:match[2]])
```

```python
matcher = Matcher(nlp.vocab)
pattern_q = [{'ORTH': '"'}, {'IS_ALPHA': True, "OP": "+"}, {'IS_PUNCT': True, "OP": "*"}, {'ORTH': '"'}]
matcher.add("QUOTES", [pattern_q], greedy='LONGEST')
doc = nlp(text)
matches_q = matcher(doc)
matches_q.sort(key = lambda x: x[1])
print (len(matches_q))
for match in matches_q[:10]:
    print (match, doc[match[1]:match[2]])
```
